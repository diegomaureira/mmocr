```python
from mmocr.apis import TextDetInferencer, TextRecInferencer, TextSpotInferencer
```

```python
test_image = '/home/diego/Documents/MLCDataset_v1/720419-MLC49647064176_042022.jpg'
```

```python
text_det_inferencer = TextDetInferencer(model='textsnake_resnet50-oclip_fpn-unet_1200e_ctw1500')
text_rec_inferencer = TextRecInferencer(model='svtr-base_20e_st_mj')
```

```python
text_det_vis = text_det_inferencer(test_image, return_vis=True)
```

```python
text_rec_vis = text_rec_inferencer(test_image, return_vis=True)
```

```python
import matplotlib.pyplot as plt
plt.imshow(text_det_vis['visualization'][0])
plt.show()
plt.imshow(text_rec_vis['visualization'][0])
plt.show()
```
